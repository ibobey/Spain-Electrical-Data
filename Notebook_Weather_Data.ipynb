### Probablitiy Density Function  

```python
from numpy import e,pi,power,sqrt
```

```python
def pdf(x,mean,std):
    return power((std * sqrt(2 * pi)),-1) * power(e,(-(x-mean)**2) / (2* std **2))
```

```python
Temperature_C = df_weather.Temperature - 273
```

```python
x = np.arange(Temperature_C.min(),Temperature_C.max(),0.01)
mean = np.mean(Temperature_C)
std = np.std(Temperature_C)
y_pdf = pdf(x,mean,std)
```

```python
sns.set_theme(style="white")
sns.set_context("paper", font_scale=1.2, rc={"lines.linewidth": 2})
```

```python
plt.figure(figsize=(11.5,5.5))

plt.hist(Temperature_C,
         color= "#0096FF",
         bins=50,
         edgecolor="white",
         alpha=.75,
         density=True,
         label="Tempreature"
        )

plt.plot(x,
         y_pdf,
         color="#DD4124",
         lw=2.5,
         ls="--",
         label="PDF"
        )

plt.fill_between(x,
                 y_pdf,
                 alpha=0.2,
                 color="#75FF33")

plt.title("Temperature")
plt.xlabel("Celcius")
plt.ylabel("Density")

sns.despine()

plt.legend(loc='best')
plt.grid(True,alpha=0.65,ls="-.")
plt.tight_layout()
plt.show()
```

```python
data_temperature = df_weather[
    (df_weather.Time.dt.strftime("%Y-%m-%d") > '2018-05-01') &
    (df_weather.Time.dt.strftime("%Y-%m-%d") < '2018-08-30')
          ].groupby(df_weather.Time.dt.date)['Temperature'].mean() - 273 # Kelvin to Celcius 
```

```python
data_temperature = data_temperature.reset_index()
```

```python
x = list(data_temperature.index)
y = list(data_temperature.Temperature)

b, a = np.polyfit(x, y, deg=1)
seq = np.linspace(min(x), max(x), num=100)
```

```python
sns.set_theme(style="white")
sns.set_context("paper", font_scale=1.2, rc={"lines.linewidth": 2})
```

```python
plt.figure(figsize=(11.5,5.5))

plt.plot(seq, 
         a + b * seq,
         c="b",
         ls="-.",
         lw=2,
         #alpha=0.8,
         label="LinReg")

plt.scatter(x,
            y,
            s=25,
            c='r',
            alpha=0.85,
            label="Temperature")

plt.xlabel("day")
plt.ylabel("Temperature (C)")
plt.ylim(min(y) - 5 ,max(y) + 5)
plt.title("Summer Temperature Regression")

sns.despine()
plt.legend(loc=2)
plt.grid(True,alpha=0.7,ls="-.")

plt.tight_layout()
plt.show()
```
